```python
import matplotlib.pyplot as plt
import numpy as np

x = [1, 2, 3, 4, 5]  # Prices
y = [5, 4, 3, 2, 1]  # Demand

plt.plot(x, y)
plt.xlabel('Prices')
plt.ylabel('Demand')
plt.title('Product Demand vs Price')
plt.savefig('prices_demand_plot.png')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]  # Age
y = [75, 80, 88, 95, 105]  # Height in cm

plt.scatter(x, y, marker='o')
plt.xlabel('Age (years)')
plt.ylabel('Height (cm)')
plt.title('Children\'s Height vs Age')
plt.savefig('age_height_plot.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

ax1.plot(x, y1, color='tab:blue')
ax1.set_title('y = sin(x)')

ax2.scatter(x, y2, marker='*', color='pink')
ax2.set_title('y = cos(x)')

plt.show()
```

```python
x = np.linspace(-10, 10, 1000)
y1 = np.sin(x)
y2 = np.cos(x)

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(7, 4))

ax1.plot(x, y1, color='tab:blue', label='y = sin(x)')
ax1.legend()

ax2.scatter(x, y2, marker='*', color='pink', label='y = cos(x)')
ax2.legend()

plt.show()
```
